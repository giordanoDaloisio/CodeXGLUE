```python
import pandas as pd
import os
```

```python
times = pd.DataFrame()
for file in os.listdir('.'):
    if file.endswith('.csv'):
        df = pd.read_csv(file)
        df = df.groupby('task_id')['time_per_sample'].mean().reset_index()
        approach = file.split('_')[1]
        times[approach] = df['time_per_sample']
```
