```python
import pandas as pd
import os
```

```python
cpu_times = pd.DataFrame()
for file in os.listdir():
    if file.endswith('.csv'):
            approach = file.split('_')[-1].split('.')[0]
            df = pd.read_csv(file, header=None)
            cpu_times[approach] = df.values.flatten()
```

```python
cpu_times.dropna(inplace=True)
```

```python
cpu_times.to_csv('../../../../../analysis/inf_times_cuda_ct_llama.csv', index=False)
```
