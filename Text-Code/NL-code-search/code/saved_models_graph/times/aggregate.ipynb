```python
import pandas as pd
import os
```

```python
cpu_times = pd.DataFrame()
for file in os.listdir():
    if file.endswith('.csv'):
        if 'cpu' in file:
            approach = file.split('_')[0]
            df = pd.read_csv(file, header=None)
            cpu_times[approach] = df.values.flatten()
```

```python
gpu_times = pd.DataFrame()
for file in os.listdir():
    if file.endswith('.csv'):
        if 'cuda' in file:
            approach = file.split('_')[0]
            df = pd.read_csv(file, header=None)
            gpu_times[approach] = df.values.flatten()
```

```python
gpu_times.to_csv('../../../../../analysis/inf_times_cuda_tc_graph.csv', index=False)
```
